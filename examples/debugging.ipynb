```python
import timecorr as tc
import ipyvolume
import numpy as np
from scipy.linalg import toeplitz
from scipy.spatial.distance import squareform, cdist, pdist
import seaborn as sns
from matplotlib import pyplot as plt
import hypertools as hyp
from ipywidgets import FloatSlider, ColorPicker, VBox, jslink

%matplotlib inline
```

```python
K = toeplitz(np.arange(1, -0.1, -0.1))
sK = tc.mat2vec(K)
X = tc.simulate.generate_data(sK, T=1000)
```

```python
sns.heatmap(tc.mexican_hat_weights(100, params={'sigma': 10}))
```

```python
corrs = tc.timecorr(X, weights_function=tc.laplace_weights)
sns.heatmap(corrs)
```

```python
ipyvolume.clear()
ipyvolume.quickvolshow(tc.vec2mat(corrs), lighting=True, controls=True)
```

```python
sns.heatmap(np.squeeze(tc.vec2mat(corrs)[:, :, 100]))
```
