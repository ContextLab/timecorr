```python
import timecorr as tc
import ipyvolume
import numpy as np
from scipy.linalg import toeplitz
from scipy.spatial.distance import squareform, cdist, pdist
import seaborn as sns
from matplotlib import pyplot as plt
import hypertools as hyp

%matplotlib inline
```

```python
K = toeplitz(np.arange(1, -0.1, -0.1))
sK = tc.mat2vec(K)
X = tc.simulate.generate_data(sK, T=1000)
```

```python
corrs = tc.timecorr(X, weights_function=tc., weights_params=None)
```

```python
sns.heatmap(corrs)
```

```python
ave_corrs = 1 - squareform(pdist(X.T, 'correlation'))
```

```python
ipyvolume.volshow(tc.vec2mat(corrs), lighting=True, level=np.linspace(0, 1, 11))
```

```python
sns.heatmap(np.squeeze(tc.vec2mat(corrs)[:, :, 100]))
```
