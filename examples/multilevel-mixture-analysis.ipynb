```python
f, ax1 = plt.subplots(1,4, sharey='row', figsize=(20,5))
plt.suptitle("pieman dataset multilevel mixture analysis",fontsize=15)

ax1[0].set_title("intact")
ax1[1].set_title("paragraph")
ax1[2].set_title("word")
ax1[3].set_title("resting")
ax1[0].set_ylabel("decoding accuracy")

data = np.load("../data/multilevel_mixture_analysis/intact.npz")["arr_1"].T
df = pd.DataFrame(data,columns = ['mix']+range(11))
newplot = sns.violinplot(data = df,orient = 'v',ax = ax1[0])
ax1[0].set_xlabel("level")

data = np.load("../data/multilevel_mixture_analysis/paragraph.npz")["arr_1"].T
df = pd.DataFrame(data,columns = ['mix']+range(11))
newplot = sns.violinplot(data = df,orient = 'v',ax = ax1[1])
ax1[1].set_xlabel("level")

data = np.load("../data/multilevel_mixture_analysis/word.npz")["arr_1"].T
df = pd.DataFrame(data,columns = ['mix']+range(11))
newplot = sns.violinplot(data = df,orient = 'v',ax = ax1[2])
ax1[2].set_xlabel("level")

data = np.load("../data/multilevel_mixture_analysis/resting.npz")["arr_1"].T
df = pd.DataFrame(data,columns = ['mix']+range(11))
newplot = sns.violinplot(data = df,orient = 'v',ax = ax1[3])
ax1[3].set_xlabel("level")

plt.ylim()
plt.show()
```
